##  Identify Outliers with 1.5IQR Method to Replace 


```python
Q1 = np.percentile(vTargetBuyers['TotalChildren'], 25)
Q3 = np.percentile(vTargetBuyers['TotalChildren'], 75)
IQR = Q3 - Q1
IQR = 1.5 * IQR
L = Q1 - IQR
h = Q3 + IQR
print("(Q1):", Q1, "Q3:", Q3,"IQR:", IQR)
print("lfence", L ,"hfence", h)

```

## Discretization (Binning/Histogram) on Continuous features or Numeric features 
with too many different values 

```python

df = pd.DataFrame(vTargetBuyers) 
doller_labels = ['< 50K', '50K - 75K', '75K - 100K', '100K+']
df['IncomeGroup'] = pd.cut(df['YearlyIncome'], bins=[0, 50000, 75000, 100000, 150000], labels=doller_labels)
df.drop('YearlyIncome', axis=1, inplace=True)
df
```

```python
df = pd.DataFrame(vTargetBuyers) 
c_labels = ['None', '1 Child', '2 Children', '3 Children', '4 Children','5 children']
df['child'] = pd.cut(df['NumberChildrenAtHome'], bins= [-1, 0, 1, 2, 3, 4, 5] , labels=c_labels)
df.drop('NumberChildrenAtHome', axis=1, inplace=True)
df
```

```python
df = pd.DataFrame(vTargetBuyers) 
c_labels = ['None', '1 Chd', '2 Chd', '3 Chd', '4 Chd','5 chd']
df['childtot'] = pd.cut(df['TotalChildren'], bins= [-1, 0, 1, 2, 3, 4, 5] , labels=c_labels)
df.drop('TotalChildren', axis=1, inplace=True)
df
```

```python
df = pd.DataFrame(vTargetBuyers) 





df['Agegroup'] = pd.cut(df['Age'], bins= [0.0, 20.0, 30.0, 40.0, 50.0, 60.0, 100.0] , labels=['0-20', '21-30', '31-40', '41-50', '51-60', '61+'])
df.drop('Age', axis=1, inplace=True)
df


```

```python
df = pd.DataFrame(vTargetBuyers) 
cars_labels = ['None', '1 car', '2 Car', '3 Car', '4 Car','5 car']
df['car'] = pd.cut(df['NumberCarsOwned'], bins= [-1, 0, 1, 2, 3, 4, 5], labels=cars_labels)
df.drop('NumberCarsOwned', axis=1, inplace=True)
df
```

## Normalization or Standardization 


```python
mean_income = vTargetBuyers['YearlyIncome'].mean()
std_income = vTargetBuyers.std()['YearlyIncome']
vTargetBuyers['YearlyIncome'] = (vTargetBuyers['YearlyIncome'] - mean_income) / std_income
print(vTargetBuyers['YearlyIncome'])
```

```python
mean_age=  vTargetBuyers['Age'].mean()

std_age= vTargetBuyers.std()['Age']

vTargetBuyers['Age'] = (vTargetBuyers['Age'] - mean_age) / std_age
print(vTargetBuyers['Age'])
```
